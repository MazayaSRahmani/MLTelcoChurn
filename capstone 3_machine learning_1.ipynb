# 1. Business Problem Understanding

### **Context**

Telecom is the #1 industry hardest hit by churn. According to TechSee report [2022 State of Customer Loyalty and Churn in Telecom](https://techsee.me/wp-content/uploads/2022/09/2022-Churn-Survey-Telecom.pdf) Survey online, the telecom industry is especially vulnurable to churn from poor service quality which is evident by 46% of those surveyed in the report canceling a phone, internet, TV or cable contract with a company in the past year.Not mention, about 92% of telecom customers that are unhappy with the sevice quality would churned to a competitior. Even 48% of churned customers are actively searched for a competitor and 44% did not feel any loyalty to the brand and would switch to a competitor after being approached.

&nbsp;
> Then why is customer churn important? To be exact, **shouldn't the company just acquired more customers and forget those who are not loyal?**

The answer is; it's way more expensive to acquire new customer in comparison to just retain customers that the company already had. According to [Forbes](https://www.forbes.com/sites/forbestechcouncil/2020/10/30/acquiring-subscribers-is-only-half-the-battle/?sh=12e28bae1821) in 2020, acquiring subscribers is only half the battle. It costs CSPs(Communication Service Providers) or a Telecom Company $315 per new subscriber. Which makes the Telecom as one of **the most expensive industries** when it comes to customer acquisition costs (CAC).

Additionally [Corporate Finance Institute](https://corporatefinanceinstitute.com/resources/valuation/cac-ltv-ratio/) stated that, a healthy Lifetime Value (LTV) to CAC ratio should be greater than 3-to-1. This means The amount of Lifetime Value that a Telecom company recieve from a customer should be 3 times higher than the Customer Acquisition Costs.But the matter of fact is that according to [LightReading](https://www.lightreading.com/broadband/us-broadband-subscriber-slowdown-could-spark-arpu-gains) a customer's average revenue per user (ARPU) is $76 per month for one of the largest telecommuncations and mass media companies [Charter Communications](https://www.statista.com/statistics/624145/charter-arpu/), this means that it will take more than a year of subscription payments to achieve the desired LTV-to-CAC ratio.

Of course this make sense as by acquiring new customers, you'd have to spend more time and marketing expenses on potential customers that has no guarantee of actually being a customer. On the other hand, by retaining the already existing ones, the company would already have the data and information of the customers to make an informed decision on what campaign should be deploy for each customer group. This statement is also echoed by research done by [Frederick Reichheld of Bain and Company](https://media.bain.com/Images/BB_Prescription_cutting_costs.pdf), where the study shows that increasing customer retention rates by 5% increases profits by 25% to 95%.

### **Problem Statement**


> So what exactly is **churn rate**?

According to Jill Avery on [Harvard Business Review](https://hbr.org/2014/10/the-value-of-keeping-the-right-customers), customer churn rate is a metric that measures the precentage of customers who end their relationship with a company in a particular period.

&nbsp;
> How do you **calculate** it?

Since churn rate is simply the percent of customers who end their relationship with your company in a given period, calculating it is pretty straightforward. You take the total number of customers who left your company during that period divided by total customers at the beginning of the period. As you can see, this is a lagging indicator, meaning you can only look at what’s happened, which is one of the metric’s downsides.

&nbsp;
> Most important of all, **how would a machine learning model can help** in this situation?

This mean that machine learning would help in this situation because churn is a lagging indicator, which meant a company can only analyze things when it already happened, machine learning would simply avoid the problem in the first place without it happening first. 

Even if the machine have some errors in predicting the churn customers, the focus of this model should be on minimizing the cost. Which in the model would mean to prioritize correctly predicting the churn customer since the cost of retaining them would be more affordable. In this case then, there would be more capital-cushion.


### **Goals**
In this project our goal is to predict the customers that are more likely to exit or churn to then maximize customer retention and minimize the cost of acquiring new customers. By that definition we would need these goals to be achieved by the models:

1. Predict the most probable customer to churn
2. The result of the prediction must save company's money 
3. Recommend the best way to intervene before a customer quit the company's telco service




### **Metrics Evaluation**
With the stated goals above, this means that it is way more important to predict the customers that are more likely to churn, which will make churn customers as our positive class and the not churn customers as the negative class.

Class categorization:
- Churn : Positive class
- Not Churn : Negative class

Metrics:
With the catgorization above, now we can sort out what is important to us as a business that wants to prioritze our retention rate. In this case we would have to focus on minimizing the amount of False Negative in the our model prediction. To get the point accross, when we focus on the **False Negative** (customers who are predicted to not churn but in actuality, they are)

It is better to miss-predict customers who are going to churn but in actuality they are not, than it is to miss-predict customers who are not going to churn but in actuallity they are. 

This means that in a worst case scenario, if we are miss-predicting people that they are going to churn, probably the company would have to spend more on marketing campaigns on the already customers.

In the other hand, the latter scenario would make the company have to spend more capital on acquiring new customers based on the given context above. Since the model would miss-predict the customers who are not going to churn but in actuality they are, the company would miss on maximizing the customers they already had and would end up needing to acquire new customers.



### Evaluation Metrics:

1. **Customer Retention Cost (CRC)**: The cost of retaining an existing customer through marketing campaigns, discounts, and loyalty programs.
  
2. **Customer Acquisition Cost (CAC)**: The cost of acquiring a new customer through advertising, promotions, and sales efforts.

3. **Customer Lifetime Value (LTV)**: The total revenue expected from a customer over their entire relationship with the business, accounting for gross margin and churn rate.

### Calculation and Evaluation:

#### 1. Calculate CRC and CAC:
- **CRC**: Let's assume the CRC per customer is $100.
- **CAC**: Assume the CAC per customer is $500.

#### 2. Calculate LTV:
- **LTV Formula**: LTV = (ARPU * Gross Margin) / Churn Rate
  - Let's assume:
    - ARPU (Average Revenue Per User): $1,000 annually
    - Gross Margin: 70% (0.7)
    - Churn Rate: 10% (0.1)
    
  Substituting the values:
  $$
  \text{LTV} = \frac{\$1,000 \times 0.7}{0.1} = \$7,000
  $$

   


#### 3. ROI Calculation:
- **ROI on Retention**: 
  $$
  \text{ROI} = \frac{\text{LTV} - \text{CRC}}{\text{CRC}} = \frac{\$7,000 - \$100}{\$100} = \$69
  $$


  This means for every dollar spent on retention, you get $69 in return.
  
- **ROI on Acquisition**: 
  $$
  \text{ROI} = \frac{\text{LTV} - \text{CAC}}{\text{CAC}} = \frac{\$7,000 - \$500}{\$500} = \$13
  $$
  This means for every dollar spent on acquisition, you get $13 in return.

#### 4. Strategic Implications:
- Given the high ROI on retention (69) compared to acquisition (13), it's financially more beneficial to focus on retention efforts.
- An acceptable false positive rate in churn prediction should be determined based on the net benefit achieved, considering the costs of false positives against the savings from retention efforts.

#### 5. Conclusion:
We're going to use f2 metrics so that we can prioritze recall without neglecting precision. In the implementation, by prioritizing recall we are saving money since falsely predictin gchurn customer who inactuality is not churn still saves money. In the other hand we also don't want to get too low of precision so that we're not spending unnecessary expenditure on marketing or retention efforts with false target.

> #### Checking Data Distributions with shapiro wilk:

> #### Checking Outliers

Because the data is assumed not distributed normally, Interval Quartile Range or IQR is used to see the percentage of the outliers out of all the data

```python
def outlier_percentage(data,column):
    Q1 = data[column].quantile(1/4)
    Q3 = data[column].quantile(3/4)
    iqr = Q3-Q1
    outlier_lower_limit = Q1 - 1.5 * iqr
    outlier_upper_limit = Q3 + 1.5 * iqr
    num_of_outliers = len(data[(data[column]<outlier_lower_limit) | (data[column]>outlier_upper_limit)])
    outlier_percentage = round(num_of_outliers/len(data[column])*100,2)
    return num_of_outliers, outlier_percentage, outlier_lower_limit, outlier_upper_limit
    
outliers = []
   
for i in numerical:
    total_outliers, outliers_percent, lower_bound, upper_bound = outlier_percentage(data,i)
    outliers.append([i, total_outliers, outliers_percent, lower_bound, upper_bound])
    
outliers_info = pd.DataFrame(outliers, columns = ["Columns", "Number of Outliers","(%) of Outliers","Lower bound","Upper bound"])
outliers_info
```

<u>**Insights**</u>:

- Both variables, `tenure` and `MonthlyCharges`, have 0 outliers detected. This indicates that all data points for these variables fall within the expected range, as defined by the lower and upper bounds. This means there's an absence of outliers.

- Since the negative lower bounds are a byproduct of the statistical method used for outlier detection and do not reflect the actual values in the dataset. Additionally, because there are no negative values in the `tenure` and `MonthlyCharges` columns, the data is clean and free from outliers, indicating high data quality and reliability for subsequent analysis.

> Chi-square test

```python
from scipy.stats import chi2_contingency

def chi2_calculator(col1, col2):
    alpha = 0.05
    contingency_table = pd.crosstab(col1, col2)
    stat, pvalue, dof, expected_freq = chi2_contingency(contingency_table)
    if pvalue > alpha:
        result = "Fail to Reject Ho"
        judgement = "Independent"
    else:
        result = "Reject Ho"
        judgement = "Dependent"
    return result, pvalue, judgement

results = []
judgement = []

plot_columns = ['Dependents', 'OnlineSecurity', 'OnlineBackup',
                'InternetService', 'DeviceProtection', 'TechSupport',
                'Contract', 'PaperlessBilling']

for i in plot_columns:
    result, pvalue,judgement = chi2_calculator(data['Churn'], data[i])
    results.append({'column': i, 'pvalue': pvalue, 'result': result,'verdict': judgement})

results_df = pd.DataFrame(results)
results_df
```

<u>**Insights**</u>:

- Based on the Chi-squared test results, with p-values consistently below the alpha, there is sufficient evidence to reject the null hypothesis (Ho) in all tests conducted. This indicates that there is a significant association between the pairs of categorical variables tested.

- In other words each categorical columns are dependent to churn

- Notably, features such as "Contract" and "InternetService" exhibit particularly low p-values (1.136356e-189 and 1.869800e-112), indicating strong associations with other categorical features. These features might be important predictors of other variables in the dataset and could have a significant impact on model performance in predictive modeling tasks.

### Splitting Data into Train and Test set

```python
# data splitting

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0, stratify=y
)
```

- Here we split the data into train and test split so we're able to train the model using the train set and then see how it perform by using the test set

### Making Column Transformer

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import KBinsDiscretizer

preprocessor = ColumnTransformer([
    ('discretizerTenure', KBinsDiscretizer(n_bins=6, strategy='uniform', encode='ordinal'), ['tenure']),
    ('discretizerMCharges', KBinsDiscretizer(n_bins=10, strategy='kmeans', encode='ordinal'), ['MonthlyCharges']),
    ('oneHot', OneHotEncoder(drop='first'), ['Dependents', 'OnlineSecurity', 'OnlineBackup','InternetService',
                                             'DeviceProtection', 'TechSupport', 'Contract','PaperlessBilling'])
], remainder='passthrough')

X_train_prep = preprocessor.fit_transform(X_train)
X_test_prep = preprocessor.transform(X_test)
```

- Here we encode all of the categorical features, and since the categorical features have low cardinality we're going to use OneHot encoder
- We're also going to bin or group the `tenure, MonthlyCharges` so that we will have a range of approximation
- This process  will turn the train set from having 10 features or columnn into 24 columns
- In the meantime i'm just not going to use `drop='first'` to make it easier to interpret for me personally


```python
feature_names = preprocessor.get_feature_names_out()
X_train_prep_df = pd.DataFrame(X_train_prep, columns=feature_names)
X_test_prep_df = pd.DataFrame(X_test_prep, columns=feature_names)
```

### Make scorer

```python
from sklearn.metrics import fbeta_score, make_scorer

# Create scorer
f2_score = make_scorer(fbeta_score, beta=2)
```

### Defining Different Resampling Options

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline as ImbPipeline

# Resampler
smote = SMOTE(random_state=0)
rus = RandomUnderSampler(random_state=0)
ros = RandomOverSampler(random_state=0)


resamplers = [rus, ros, smote]
```

### Defining Models

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from xgboost.sklearn import XGBClassifier

# Define the models
logreg = LogisticRegression(solver='liblinear', random_state=0)
knn = KNeighborsClassifier(n_neighbors=5)
dtree = DecisionTreeClassifier(criterion='entropy', min_samples_leaf=1, max_depth=10, random_state=0)
rf = RandomForestClassifier(random_state=0)
adaboost = AdaBoostClassifier(random_state=0)
gradboost = GradientBoostingClassifier(random_state=0)
xgboost = XGBClassifier(random_state=0, verbosity=0)
```

- Model benchmarking

    Here we're just going to see which models would be the best fit for our prediction by seeing the f2 score, note that here we're just doing the default parameters to speed things up. From this point on we're going to see the top performers and then tune it better one by one

```python
# Pilihan model
models = [logreg, knn, dtree, rf, adaboost, gradboost,xgboost]
resamplers = [rus, ros, smote]


f2_mean = []
f2_std = []
all_f2 = []
model_name = []
resampler_name = []

# Mencari model terbaik berdasarkan rerata f2 score terkecil dan std terkecil
for algoritma in models:
    for resampler in resamplers:

        model_pipe = ImbPipeline([
            ('preprocessing', preprocessor), 
            ('resampler', resampler),
            ('model', algoritma)
        ])

        skfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

        model_cv = cross_val_score(
            model_pipe,                 # model yang sudah dimasukkan ke dalam pipeline
            X_train,                    # data sebelum di preprocessing
            y_train, 
            cv = skfold, 
            scoring = f2_score,
            error_score='raise'
            )

        f2_mean.append(model_cv.mean())
        f2_std.append(model_cv.std())
        all_f2.append(model_cv.round(4))
        model_name.append(algoritma.__class__.__name__)
        resampler_name.append(resampler.__class__.__name__)
        
        
cv_results_df = pd.DataFrame({
    'model': model_name,
    'resampler': resampler_name,
    'mean f2 (cv)': f2_mean,
    'std': f2_std
}).sort_values(['mean f2 (cv)','std'], ascending=False)
cv_results_df.head()
```

```python
# Evaluate the models on the test set
test_f2_scores = []
model_names_test = []
resampler_names_test = []

for algoritma in models:
    for resampler in resamplers:
        model_pipe = ImbPipeline([
            ('preprocessor', preprocessor),
            ('resampler', resampler),
            ('model', algoritma)
        ])

        # Fitting the model
        model_pipe.fit(X_train, y_train)

        # Predicting
        y_pred = model_pipe.predict(X_test)

        # Evaluate with F2 score
        test_f2 = fbeta_score(y_test, y_pred, beta=2)
        test_f2_scores.append(test_f2)
        model_names_test.append(algoritma.__class__.__name__)
        resampler_names_test.append(resampler.__class__.__name__)

# Store test set results in a DataFrame
test_results_df = pd.DataFrame({
    'model': model_names_test,
    'resampler': resampler_names_test,
    'f2 (test_set)': test_f2_scores
}).sort_values('f2 (test_set)', ascending=False)


test_results_df.head()
```

> Comparing between the cross validation results and the testing results

- The differences (diff column) range from 0.000474 to 0.018547, suggesting slight improvements in test set performance compared to cross-validation results. This is expected to some extent due to variance in the data and the randomness of the split, even with a fixed random state.

- Generalization Ability: The models seem to generalize well from the training data to the unseen test data, as indicated by the small differences in F2 scores.

The slightly higher test set scores could be attributed to normal variations due to random sampling. Overall, the models are robust, and the resampling techniques improve their ability to handle imbalanced data.

```python
# define the features and target
X = data.drop(columns=['Churn','Target','MonthlyChargesGroup',
                       'tenureYear','DeviceProtection','MonthlyCharges',
                       'Dependents','TechSupport','OnlineBackup'])
y = data['Target']

# data splitting
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0, stratify=y
)

preprocessor = ColumnTransformer([
    ('discretizerTenure', KBinsDiscretizer(n_bins=6, strategy='uniform', encode='ordinal'), ['tenure']),
    ('oneHot', OneHotEncoder(drop='first'), ['InternetService','Contract','PaperlessBilling','OnlineSecurity'])
], remainder='passthrough')

X_train_prep = preprocessor.fit_transform(X_train)
X_test_prep = preprocessor.transform(X_test)
```

```python
# Pilihan model
models = [logreg, knn, dtree, rf, adaboost, gradboost,xgboost]
resamplers = [rus, ros, smote]


f2_mean = []
f2_std = []
all_f2 = []
model_name = []
resampler_name = []

# Mencari model terbaik berdasarkan rerata f2 score terkecil dan std terkecil
for algoritma in models:
    for resampler in resamplers:

        model_pipe = ImbPipeline([
            ('preprocessing', preprocessor), 
            ('resampler', resampler),
            ('model', algoritma)
        ])

        skfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

        model_cv = cross_val_score(
            model_pipe,                 # model yang sudah dimasukkan ke dalam pipeline
            X_train,                    # data sebelum di preprocessing
            y_train, 
            cv = skfold, 
            scoring = f2_scorer,
            error_score='raise'
            )

        f2_mean.append(model_cv.mean())
        f2_std.append(model_cv.std())
        all_f2.append(model_cv.round(4))
        model_name.append(algoritma.__class__.__name__)
        resampler_name.append(resampler.__class__.__name__)
        
        
cv_results_df = pd.DataFrame({
    'model': model_name,
    'resampler': resampler_name,
    'mean f2 (cv)': f2_mean,
    'std': f2_std
}).sort_values(['mean f2 (cv)','std'], ascending=False)
cv_results_df.head()
```

```python
# Evaluate the models on the test set
test_f2_scores = []
model_names_test = []
resampler_names_test = []

for algoritma in models:
    for resampler in resamplers:
        model_pipe = ImbPipeline([
            ('preprocessor', preprocessor),
            ('resampler', resampler),
            ('model', algoritma)
        ])

        # Fitting the model
        model_pipe.fit(X_train, y_train)

        # Predicting
        y_pred = model_pipe.predict(X_test)

        # Evaluate with F2 score
        test_f2 = fbeta_score(y_test, y_pred, beta=2)
        test_f2_scores.append(test_f2)
        model_names_test.append(algoritma.__class__.__name__)
        resampler_names_test.append(resampler.__class__.__name__)

# Store test set results in a DataFrame
test_results_df = pd.DataFrame({
    'model': model_names_test,
    'resampler': resampler_names_test,
    'f2 (test_set)': test_f2_scores
}).sort_values('f2 (test_set)', ascending=False)


test_results_df.head()
```

### Model Parameters

1. **Criterion: 'gini'**:
   - The Gini impurity criterion is used to measure the quality of a split. The model aims to reduce impurity in each split, which means it tries to make each node as pure as possible (having mostly one class).

2. **Max Depth: 10**:
   - The maximum depth of the tree is limited to 10 levels. This constraint helps prevent the model from becoming too complex and overfitting the training data. It strikes a balance between complexity and generalization.

3. **Max Features: 'sqrt'**:
   - At each split, the model considers a subset of features equal to the square root of the total number of features. This approach helps in reducing overfitting by adding randomness to the model and improving its ability to generalize.

4. **Min Samples Leaf: 4**:
   - A minimum of 4 samples are required to be at a leaf node. This parameter prevents the model from creating nodes that are too specific and ensures that each leaf node represents a sufficiently large subset of the data.

5. **Min Samples Split: 10**:
   - A minimum of 10 samples are required to split an internal node. This parameter ensures that splits are only made when there is a significant number of samples, helping to prevent overfitting.
